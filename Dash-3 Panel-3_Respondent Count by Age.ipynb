```python
import pandas as pd
from pathlib import Path
df = pd.read_csv('E:\Courses\COURSERA\IBM Data Analyst Professional Certificate\IBM Data Analyst Capstone Project\W-5\CSV\m5_survey_data_demographics.csv')
df
```

```python
#check if realy have respondent 99 years old!:
max_value = df['Age'].max()
print(max_value)
```

```python
age_df = df[['Age', 'Respondent']].groupby('Age').count().reset_index()
age_df['Age'] = age_df['Age'].astype(int)
age_df
```

```python
#convert data frame to cvs and save to local disk
import tempfile
tempdir = tempfile.gettempdir()
age_df.to_csv(tempdir + '/age_df.csv', index=False)
from IPython.display import FileLink
FileLink(tempdir + '/age_df.csv')
```

```python
import matplotlib.pyplot as plt
age_df.plot(x='Age', y='Respondent', kind='line')
# set the chart title and axis labels
plt.title('Number of respondents by age')
plt.xlabel('Age')
plt.ylabel('Number of respondents')

# show the chart
plt.show()
```
